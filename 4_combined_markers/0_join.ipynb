```r
sources <- c("clinical", 
             "purity", 
             "drivers", 
             "cn",
             "cn_simple",
             "isofox_genesets",
             "isofox_malignant_genesets",
             "isofox_insta_mp", 
             "isofox_insta_prop",
             "teal", 
             "viral", 
             "lilac", 
             "neo", 
             "neo_pep")
```

```r
dfs <- list()
for( i in sources){
  dfs[[i]] <- fread( paste0(i, "_ready.csv"))
}
```

```r
ready <- dfs %>% reduce(left_join, by = "sampleId")
```

#### 1 - Small cleaning

```r
ready <- ready %>% mutate_at( names(ready %>% select(contains("viral_"))), ~replace_na(.,0))
```
