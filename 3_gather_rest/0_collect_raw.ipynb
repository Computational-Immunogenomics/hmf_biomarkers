#### 1 - Collect Lilac

```r
LILAC_DIR <- paste0(I_DIR, 'lilac/lilac_out/')
lilac_patients <- unique(unlist(lapply(list.files(LILAC_DIR), function(i) strsplit(i, ".lilac")[[1]][1])))
```

```r
lilac <- list(); 
lilac_qc <- list()
system.time(
for( i in lilac_patients){
  i_file <- get_fp(i, type = "lilac")
  if(file.exists(i_file)){ 
    lilac[[i]] <- reader( i_file, sample = i) 
  }
  i_file <- get_fp(i, type = "lilac_qc")  
  if(file.exists(i_file)){ 
    lilac_qc[[i]] <- reader( i_file, sample = i) 
  }   
})
```

```r
fwrite( do.call("rbind", lilac), paste0(TMP_DIR, "lilac.csv"))
fwrite( do.call("rbind", lilac_qc), paste0(TMP_DIR, "lilac_qc.csv"))
```

#### 3 - Collect Neoepitopes

```r
NEO_DIR <- paste0(I_DIR, 'neo/data_neo/')
neo_patients <- list.files(NEO_DIR)
```

```r
neo <- list(); neo_pep <- list()
system.time(
for( i in neo_patients ){
  i_file <- get_fp(i, type = "neo")
  if(file.exists(i_file)){    
      neo[[i]] <- reader( i_file = get_fp(i, type = "neo"), sample = i)
  }
  i_file <- get_fp(i, type = "neo_pep")
  if(file.exists(i_file)){    
      neo_pep[[i]] <- reader( i_file, sample = i ) %>% arrange(desc(Score)) %>% head(100)
  }
})
```

```r
fwrite( do.call("rbind", neo), paste0(TMP_DIR, "neo.csv"))
fwrite( do.call("rbind", neo_pep), paste0(TMP_DIR, "neo_pep.csv"))
```
