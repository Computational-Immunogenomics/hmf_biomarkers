#### 0 - Read collected NEO data

- Neo data, just count?
- What else to extract?

```r
neo <- fread( paste0(TMP_DIR, "neo.csv"))
```

```r
head(neo)
```

```r
neo_ready <- 
neo |>
  group_by( sampleId ) |>
  summarise(neo_ct = n()) |> 
  arrange(desc(neo_ct))
```

- Almost a million peptides :|

```r
neo_pep <- fread( paste0(TMP_DIR, "neo_pep.csv"))
```

```r
neo_pep
```

```r
top_peps <- 
neo_pep |> 
  group_by(Peptide) |> 
  summarise(ct = n()) |>
  arrange(desc(ct)) |> 
  filter(ct > 30 ) |>
  pull(Peptide)
```

```r
neo_pep_ready <- 
neo_pep |>
  filter(Peptide %in% top_peps) |>
  group_by(sampleId, Peptide) |>
  mutate( rk = row_number(desc(Score))) |>
  filter( rk == 1) |>
  select(sampleId, Peptide, Score) |>
  spread(Peptide, Score)
```

```r
neo_pep_ready[is.na(neo_pep_ready)] <- 0
```

```r
colnames(neo_pep_ready) <- c("sampleId", paste0("neo_pep_", colnames(neo_pep_ready)[-1]))
```
