### 0 - Read and combined prepared datasets

```r
setwd(READY_DIR)
```

```r
sources <- 
c("clinical", 
  "purity", 
  "drivers", 
  "cn_simple",
  "isofox_genesets",
  "cider_dna", 
  "teal", 
  "viral", 
  "lilac", 
  "neo", 
  "neo_pep", 
  "svs", 
  "gie", 
  "fusions_dna")
```

```r
dfs <- list()
system.time(
for( i in sources){
  print(i); flush.console()
  dfs[[i]] <- fread( paste0(i, "_ready.csv"))
})
```

```r
ready <- dfs %>% reduce(left_join, by = "sampleId") 
```

```r
names(ready) <- unlist(lapply(names(ready), function(i) gsub( "[ -]", "_", i)))
```

### 1 - Small cleaning 

```r
ready <- ready %>% mutate_at( names(ready %>% select(contains("viral_"))), ~replace_na(.,0))
ready <- ready %>% mutate_at( names(ready %>% select(contains("driver"))), ~replace_na(.,0))
```

### 2 - Save combined dataset

```r
fwrite( ready, paste0(SHARE_DIR, "biomarkers_base.csv") )
```
