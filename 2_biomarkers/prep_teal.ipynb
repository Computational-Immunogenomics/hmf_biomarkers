# 0 - Read and process collected TEAL data

```r
teal <- fread(paste0(TMP_DIR, "teal.csv")) %>% se(sampleId, type, rawTelomereLength, finalTelomereLength)
```

```r
teal_ready <- 
ij(
 teal %>% fi(type == "ref") %>%   tm(sampleId = gsub("R", "T", sampleId), teal_ref_raw =   rawTelomereLength, teal_ref_final =   finalTelomereLength), 
 teal %>% fi(type == "tumor") %>% tm(sampleId, teal_tumor_raw = rawTelomereLength, teal_tumor_final = finalTelomereLength), 
 by = "sampleId") %>%
tm( sampleId, 
    teal_ref_raw,
    teal_tumor_raw, 
    teal_tumor_final, 
    teal_raw_ratio = log2(teal_tumor_raw/teal_ref_raw+1),
    teal_final_ratio = log2(teal_tumor_final/teal_ref_raw+1)
  ) 
```

# 1 - Send it

```r
fwrite(teal_ready, paste0(READY_DIR, "teal_ready.csv"))
```
