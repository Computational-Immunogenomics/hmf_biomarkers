# 1 - Fusions

```r
fusions <- 
fread( paste0(TMP_DIR, "structural_variants/fusion.csv")) %>% 
 fi(reported) %>% 
 tm(sampleId, gene = name, method = "FUSION") %>% 
 rw() %>% mu(upstream = strsplit(gene, "_")[[1]][1], downstream = strsplit(gene, "_")[[1]][2]) %>% ug()
```
