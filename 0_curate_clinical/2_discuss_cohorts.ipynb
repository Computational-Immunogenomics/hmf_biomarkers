### 0 - Read in data to define cohorts

```r
cohorts_ready <- fread( paste0(META_DIR, "cohorts/cohorts_ready.csv") )
```

```r
options(repr.plot.width = 12, repr.plot.height = 6)
```

```r
base <- 
cohorts_ready %>% 
# fi(primaryTumorLocation == "Breast") %>% 
 mu(same = ifelse(primaryTumorLocation %in% c("Breast", "Lung", "Soft tissue", "Skin", "Bladder", "Pancreas"),
                  "Location + Granular Definition", "Primary Tumor Location Based")) %>% 
 mu(colors = ifelse(same == "Primary Tumor Location Based", "Location Defined", primaryTumorLocation)) %>% 
 gb(primaryTumorLocation, cohort, same, colors) %>% 
 su(ct = n()) %>% 
 ar(desc(ct)) %>%
 ggplot(aes(y = fct_reorder(cohort, ct), x = ct, fill = colors)) + geom_bar(stat = "identity", color = "black") + 
 facet_wrap(~same, scales = "free") + 
 go_theme + 
 labs(title = "Cohort Definition Counts", x = "Samples", y = "") + 
 theme(legend.position = c(0.3, 0.4))

base
```

### Individual Cohorts

```r
see <- function(i) {
  cohorts_ready %>% 
   fi(primaryTumorLocation == i) %>% 
   gb(primaryTumorLocation, primaryTumorType, cohort) %>% 
   su(ct = n()) %>% 
   ar(desc(ct))
}
```

```r
see("Soft tissue") %>% fi(ct > 5)
```

```r
options(repr.plot.width = 6, repr.plot.height = 4)
```

```r
my_colors <- unique(ggplot_build(base)$data[[1]]$fill)
names(my_colors) <- c("Breast", "Lung", "Soft tissue", "Skin", "Bladder", "Pancreas", "Primary Tumor Location Based")
```

```r
plotter <- function(i = "Breast"){
cohorts_ready %>%  
 fi(primaryTumorLocation == i) %>% 
 mu(same = ifelse(primaryTumorLocation %in% c("Breast", "Lung", "Soft tissue", "Skin", "Bladder", "Pancreas"),
                  "Location + Granular Definition", "Primary Tumor Location Based")) %>% 
 mu(colors = ifelse(same == "Primary Tumor Location Based", "Location Defined", primaryTumorLocation)) %>% 
 gb(primaryTumorLocation, cohort, same, colors) %>% 
 su(ct = n()) %>% 
 ar(desc(ct)) %>%
 ggplot(aes(y = fct_reorder(cohort, ct), x = ct, fill = colors)) + geom_bar(stat = "identity", color = "black") + 
 facet_wrap(~same, scales = "free") + 
 scale_fill_manual(values = my_colors) + 
 go_theme + 
 labs(title = "Cohort Definition Counts", x = "Samples", y = "") 
}
```

```r
options(repr.plot.width = 6, repr.plot.height = 4)
```

```r
plotter("Bladder"); plotter("Soft tissue")
```
