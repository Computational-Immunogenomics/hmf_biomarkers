```r
ISO_DIR <- paste0(I_DIR, 'isofox/data_isofox/')
patients <- list.files(ISO_DIR)
```

#### 0 - Collected adjusted TPMs from files

```r
isofox_reader <- function( i_file, sample){
    fread( i_file ) |> select(GeneId, GeneName, AdjTPM)  |> setNames(c("GeneId","GeneName", sample)) 
}
```

```r
adj_tpms <- list()
system.time(
for( i in patients){
    i_file <- get_fp( i, type = "isofox" )
    if(file.exists(i_file)){
        adj_tpms[[i]] <- isofox_reader(i_file, i) 
    }
})
```

#### 1 - Join and save them!

```r
fwrite(adj_tpms |> reduce(inner_join, by = c("GeneId", "GeneName")), paste0(TMP_DIR, "isofox_adj_tmp.csv"))
```
