```r
go <- function( data = "lee.InstaPrism.res.RData" ){
    load(data)
    
    ### Format Hi-resolution data 
    step0 <- list()
    for( i in names(InstaPrism.res@map)){
        print(i)
        flush.console()
        tmp = data.frame(
          reconstruct_Z_ct_initial(
              InstaPrism_obj = InstaPrism.res, cell.type.of.interest = i)) %>% 
          rownames_to_column("gene") %>% 
          #filter(gene %in% top_genes) %>% 
          gather(sample, expression, -gene) %>% 
          mutate(cell_type = i)
        step0[[i]] <- tmp
    }
    print("Complete step 0"); flush.console()
    
    step1a <- data.table::rbindlist(step0)
    step1b <- data.table(step1a)[, .(total = sum(expression)),by=list(gene,sample)]
    step1 <- 
    step1a %>% 
      filter(cell_type == "Malignant") %>% 
      select(-cell_type) %>%
      rename(malignant = expression) %>% 
      left_join(as.data.frame(step1b), by = c("sample", "gene")) %>% 
      mutate(pct = malignant/total) 

    print("Complete step 1"); flush.console()
    ### Cell fraction estimates
    step2 <- 
    data.frame(InstaPrism.res@Post.ini.ct@theta) %>% 
      rownames_to_column("cell_type") %>% 
      gather(sample, malignant_est, -cell_type) %>% 
      filter(cell_type == "Malignant") %>% 
      select(-cell_type)
    
    print("Complete step 2"); flush.console()
    ### join and output
    step1 %>% inner_join(step2, by = "sample") %>% ungroup %>% unique()
}
```
