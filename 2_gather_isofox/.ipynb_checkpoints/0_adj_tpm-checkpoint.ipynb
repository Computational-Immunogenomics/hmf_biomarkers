```r
ISO_DIR <- paste0(I_DIR, 'isofox/data_isofox/')
```

```r
library(tidyverse)
library(data.table)
```

#### 0 - Collected adjusted TPMs from files

```r
reader <- function( i_file = paste0(ISO_DIR, "ACTN01020001T/ACTN01020001T.isf.gene_data.csv"), sample){
    fread( i_file ) %>% select(GeneId, GeneName, AdjTPM)  %>% setNames(c("GeneId","GeneName", sample)) 
}
```

```r
adj_tpms <- list()
system.time(
for( i in list.files(ISO_DIR)){
    i_file <- as.character(paste0(ISO_DIR, i, "/", i, ".isf.gene_data.csv"))
    if(file.exists(i_file)){
        print(i); flush.console()
        adj_tpms[[i]] <- reader(i_file, i)
    }
})
```

#### 1 - Join and save them!

```r
ready <- adj_tpms %>% reduce(inner_join, by = c("GeneId", "GeneName"))
```

```r
fwrite(ready, paste(TMP_DIR, "isofox_adj_tmp.csv"))
```
