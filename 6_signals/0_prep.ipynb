#### 0 - read prepped biomarkers

```r
system.time(go <- fread(paste0(SHARE_DIR, "biomarkers_ex.csv")))
```

#### 1 - Select features

- Define features for analyses

```r
clin_features <- paste0("clin_", c("purity", "sex", "age", "hasSystemicPreTreatment", "hasRadiotherapyPreTreatment"))
```

```r
features <- names( go %>% select(any_of(clin_features), 
                                 contains("cn"), 
                                 contains("driver"), 
                                 contains("lilac"), 
                                 contains("neo"), 
                                 contains("purity"), 
                                 contains("rna"), 
                                 contains("teal"), 
                                 contains("viral")))   
```

#### 2 - Prep data
- Standardize features - scaling

```r
go_scaled <- go %>% mutate_at( features, ~scale(.) ) 
```

#### 3 - Collect list of dataframes

```r
system.time(saveRDS( list( "data_ready" = go_scaled, "features" = features), paste0(SHARE_DIR, "ready_ex.Rds")))
```
