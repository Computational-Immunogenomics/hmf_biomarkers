### 3 - Derive Clinical Endpoints from Response

###### Base data to derive all outcomes
- Need response dataframe, with biopsy and death date collected from meta

```r
base <- (
    response %>% select(-name, -type, -mechanism)
             %>% left_join( meta %>% select(sampleId, biopsyDate, deathDate), by = "sampleId")
             %>% mutate( biopsyDate = ifelse( biopsyDate == "null", NA, biopsyDate),
                         deathDate = ifelse( deathDate == "null", NA, deathDate),
                         responseDate = ifelse( responseDate == "null", NA, responseDate),
                         endDate = ifelse( endDate == "null", NA, endDate))
             %>% rowwise()
             %>% mutate(lastDate = max(endDate, responseDate))
             %>% group_by( sampleId ) 
             %>% mutate(lastKnownAliveDate = max(lastDate))
             %>% ungroup()
)
```

###### Helper functions

```r
date_diff <- function(d1, d2) { ### used to compute day ranges for overall survival
    if( is.na(d1) || is.na(d2) ){ NA } 
    else { as.numeric( difftime(as.Date(d2), as.Date(d1), units = "days"))
}}
```

```r
go_bor <- function(i){ ### get best recist response from list of responses
    if( grepl("CR",i ) ){ "CR" } 
    else if (grepl("PR",i )) { "PR" } 
    else if( grepl("SD",i) ) { "SD" } 
    else if ( grepl("PD",i)) { "PD" } 
    else { NA }
}
```

```r
go_binary <- function( i ){ ### if complete or partial response than 1
    if( i %in% c("CR", "PR") ){ 1 } 
    else if ( i %in% c("SD", "PD")) { 0 } 
    else { NA } 
}
```

#### Best Overall response criteria
- Based on the recist 

```r
bor_ready <- (
  base
    %>% group_by(sampleId)
    %>% summarise( bor = go_bor(paste0( response, collapse = ",")), bor_bin = go_binary(bor))
    %>% ungroup()
)
```

#### Overall Survival
- Assume biopsy as start date
- Assume that if date of death is missing, the patient still alive
- Assume last known date alive is either last treatment date or last response date (whichever is later)

```r
os_ready <- (
  base 
    %>% rowwise() 
    %>% mutate( os_event = ifelse( is.na(deathDate), 0, 1), 
                os_days = ifelse( os_event == 1, date_diff( biopsyDate, deathDate ), date_diff(biopsyDate, lastKnownAliveDate)),
                os = ifelse(os_event == 1, os_days, -os_days))
    %>% ungroup()
    %>% select(sampleId, os, os_days, os_event)
    %>% unique()
)
```

#### Progression Free Survival
- Time to first progression or death
- Assume biopsy as start date

```r
get_pfs_days <- function( pfs_event, biopsyDate, lastKnownAliveDate, progressionDate, deathDate){
    if( pfs_event == 0){ date_diff( biopsyDate, lastKnownAliveDate)} 
    else if (!is.na(progressionDate) ){ date_diff( biopsyDate, progressionDate) }
    else if ( !is.na(deathDate) ) { date_diff( biopsyDate, deathDate) }
    else {NA}
}
```

```r
pfs_ready <- (
  base 
    %>% left_join( base %>% filter( response == "PD" ) %>% group_by( sampleId )
                        %>% summarise( progressionDate = min(responseDate)) %>% ungroup(), by = "sampleId")
    %>% mutate( pfs_event = ifelse( !is.na(progressionDate) | !is.na(deathDate), 1, 0))
    %>% rowwise()
    %>% mutate(pfs_days = get_pfs_days( pfs_event, biopsyDate, lastKnownAliveDate, progressionDate, deathDate ),
               pfs = ifelse( pfs_event == 1, pfs_days, -pfs_days))
    %>% ungroup()
    %>% select(sampleId, pfs_event, pfs_days)
    %>% unique()
)
```

##### Combined outcomes

```r
outcomes_ready <- os_ready %>% full_join( pfs_ready, by = "sampleId") %>% full_join(bor_ready, by = "sampleId")
```
