### 0 - Read and combined prepared datasets

```r
setwd(READY_DIR)
```

```r
sources <- 
c("clinical", 
  "purity", 
  "drivers", 
  "cn_simple",
  "isofox_genesets",
  "cider_dna", 
  "teal", 
  "viral", 
  "lilac", 
  "neo", 
  "neo_pep", 
  "svs", 
  "gie", 
  "fusions_dna", 
  "chord",
  "hotspots")
```

```r
dfs <- list()
system.time(
for( i in sources){
  print(i); flush.console()
  dfs[[i]] <- fread( paste0(i, "_ready.csv"))
})
```

```r
ready <- dfs %>% reduce(left_join, by = "sampleId") 
```

### 1 - Small cleaning 

- Driver and Viral should have 0's for NAs (indicate no events)

```r
ready <- 
ready %>% mutate( across(c(contains("driver"), contains("viral"), contains("hotspot")), ~replace_na(.,0) ))
```

### 2 - Save combined dataset

```r
fwrite( ready, paste0(SHARE_DIR, "biomarkers_base.csv") )
```
