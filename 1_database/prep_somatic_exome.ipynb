#### 0 - Read somatic exome

- Directory with whole somatic exome output

```r
EXOME_DIR <- paste0(TMP_DIR, "somatic_exome_big/")
```

- Remove non-translated reads (make much smaller, and easier to work with)

```r
remove <- c("intron_variant")
```

### 1 - Run the joins

```r
go <- data.frame()
```

```r
system.time(
for( i in list.files(EXOME_DIR)){
  fp <- paste0(EXOME_DIR, i)
  print(fp); flush.console()
  tmp <- fread( fp ) %>% fi(! annotation %in% remove)
  go <- rbind(go, tmp)
})
```
