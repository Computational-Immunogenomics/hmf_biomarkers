#### 1 - Compute correlations to pre-selected features

```r
cor_computer <- function( df, base, name = "prolif" ) {
  cors <- sapply(df, function(x) tryCatch(cor(df[[base]], x, use = "complete.obs"), error = function(e) NULL))
  data.frame( feature = names(unlist(cors)), correlation = unlist(cors)) %>% 
    rownames_to_column("rowname") %>% 
    select(-rowname) %>% 
    drop_na( correlation ) %>% 
    mutate( ref = base ) 
}
```
