# Prepare data for analysis

- Prepared categorical markers

```r
univariate_results <- fread(paste0(SHARE_DIR, "1_run_fishers_exact.csv")) 
fisher_base <- fread(paste0(SHARE_DIR, "fisher_base.csv"))
```

```r
example <- univariate_results %>% fi(cohortGo == "Pan-Cancer ## Anti-PD-1") 
example_base <- fisher_base %>% fi(cohortGo == "Pan-Cancer ## Anti-PD-1") 
```

- Volcano plot

```r
options(repr.plot.height =3, repr.plot.width = 3) 
```

```r
example %>% 
  ggplot(aes(x = log(or), y = -log10(fisher_pval), color = (fisher_pval < .01), alpha = (fisher_pval < .01))) + 
  geom_point() + 
  scale_alpha_manual(values = c(.005,1)) + 
  scale_color_manual(values = c("blue","orange")) + 
  geom_hline(yintercept = -log10(.01)) + 
  labs(x = "Log Odds or Response", y = "-Log10(p-value)", title = "Anti-PD1 Response Signals") + 
  go_theme + theme(legend.position = "none")
```

- Correlation Plot

```r
cor_base <- example_base %>% se(any_of(example %>% fi(fisher_pval < .001 , direction == "Non-Response") %>% pu(feature)))
```

```r
options(repr.plot.height =10, repr.plot.width = 10) 
```

```r
corrplot(cor(cor_base, use = "pairwise.complete.obs"), order = "hclust", hclust.method = "ward.D2", tl.cex = .1)
```
