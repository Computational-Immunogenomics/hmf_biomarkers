### 4 - Added Figure

```r
tmp1 <- 
data.frame( "type" = c("Non-Response", "Non-Response", "Response"),
            event = c(TRUE, FALSE, FALSE), 
            baseline = c(11, 76, 79)) %>% 
 mu("<5% Threshold" = baseline * 59/11, 
    "<2% Threshold" = baseline * 149/11) %>% 
 ga(threshold, val, -event, -type) %>% 
 mu(threshold = ifelse(threshold == "baseline", "Observed", threshold)) %>% 
 mu(Responder = factor(type, levels = rev(c("Non-Response", "Response"))),
    threshold = 
    factor(threshold, levels = c("Observed", 
                                 "<5% Threshold" , 
                                 "<2% Threshold"))) %>%
 mu( `Biomarker Event` = factor(event, levels = c(FALSE, TRUE)), example = "Skin Melanoma\nImmune Driver Evasion")
```

```r
tmp2 <- 
data.frame( "type" = c("Non-Response", "Non-Response", "Response"),
            event = c(TRUE, FALSE, FALSE), 
            baseline = c(22, 36, 21)) %>% 
 mu("<5% Threshold" = baseline * 59/22, 
    "<2% Threshold" = baseline * 149/22) %>% 
 ga(threshold, val, -event, -type) %>% 
 mu(threshold = ifelse(threshold == "baseline", "Observed", threshold)) %>% 
 mu(Responder = factor(type, levels = rev(c("Non-Response", "Response"))),
    threshold = 
    factor(threshold, levels = c("Observed", 
                                 "<5% Threshold" , 
                                 "<2% Threshold"))) %>%
 mu( `Biomarker Event` = factor(event, levels = c(FALSE, TRUE)), example = "Anti-AR\nRNA Complement/Coagulation High")
```

```r
go <- rbind(tmp1, tmp2) #%>% 
# mu(example = factor(example, levels = c("RNA Complement/Coagulation High\nAnti-AR", "Immune Driver Evasion\nSkin Melanoma")))
```

```r
tots <- go %>% ug() %>% gb(threshold, example) %>% su(total = round(sum(val))) %>% ug()
```

```r
colors <- c("Non-Response" = "#e52f28", "Response" = "#7AABD3")
```

```r
options(repr.plot.height = 6, repr.plot.width = 7.5) 
```

```r
b <- 
go %>% 
 ggplot(aes(x = threshold, y = val, fill = Responder, alpha = `Biomarker Event`, color = `Biomarker Event`)) + 
 geom_bar(stat = "identity", width = .6) + 
 schema_theme + 
 scale_fill_manual(values = unlist(colors)) + 
 scale_color_manual(values = c("white", "black")) + 
 scale_alpha_manual(values = c(.3, 1)) + 
 labs(x = "Groups", y = "Patients Observed or Needed", title = "Hypothetical Required Cohort Sizes") + 
 geom_text(data = tots, aes(x = threshold, label = total, y = total), inherit.aes = FALSE, vjust = -.4) + 
 geom_text( aes(label = round(val)), vjust = 0)
```

```r
c <- b + facet_wrap(~example, scales = "fixed") +
  theme(
    strip.background = element_blank(),     # Remove grey background
    strip.text = element_text(color = "black", size = 14)  # Set text color (optional)
  )
```
