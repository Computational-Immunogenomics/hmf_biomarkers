```r
gper <- function(i) {
 if(grepl("lt25", i)){ "low"}
 else if(grepl("gt25", i)){ "moderate"}   
 else if(grepl("50", i)){ "high"}
 else if(grepl("75", i)){ "very high"}   
}
```

```r
size_map <- list("low" = 1, "moderate" = 2, "high" = 3, "very high" = 4)
```
