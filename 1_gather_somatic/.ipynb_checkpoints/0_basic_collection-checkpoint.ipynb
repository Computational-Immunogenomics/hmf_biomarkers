```r
SOM_DIR <- paste0(I_DIR, "somatics/")
```

```r
library(tidyverse)
library(data.table)
```

#### 0 - Collect purity files across somatic data

```r
reader <- function( i_file = paste0(SOM_DIR, "ACTN01020001T/purple/ACTN01020001T.purple.purity.tsv"), sample = "ACTN01020001T"){
    fread( i_file ) %>% mutate(sampleId = sample) 
}
```

```r
purities <- list()
system.time(
for( i in list.files(SOM_DIR)){
    i_file <- as.character(paste0(SOM_DIR, i, "/purple/", i, ".purple.purity.tsv"))
    if(file.exists(i_file)){
        #print(i); flush.console()
        purities[[i]] <- reader(i_file, i)
    }
})
```

#### 1 - Bind and save them!

```r
ready <- do.call("rbind", purities)
```

```r
fwrite(ready, paste0(TMP_DIR, "purities.csv"))
```
