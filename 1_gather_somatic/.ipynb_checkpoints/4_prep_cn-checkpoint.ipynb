#### 0 - Read and process collected driver data
- Just including all drivers output from Linx here

```r
drivers <- fread( paste0(TMP_DIR, "drivers.csv"))
```

```r
drivers_ready <- 
drivers |>
  select(sampleId, gene) |>
  unique() |>
  mutate(driver = 1) |>
  spread(gene, driver)
```

```r
drivers_ready[is.na(drivers_ready)] <- 0
```

```r
cors <- cor(drivers_ready %>% select(-sampleId))
```

```r
keep <- names(sort(apply(drivers_ready[,-1],2 ,sum), decreasing = TRUE))[1:100]
```

```r
options(repr.plot.height=20, repr.plot.width=20)

corrplot(cor(drivers_ready %>% select(any_of(keep))), method = "square", order = "hclust")
```
