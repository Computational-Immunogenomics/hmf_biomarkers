#### 0 - Read and process collected CN data
- Just summarizing based on regions (lots of redundancy with genes)

```r
cn <- fread(paste0(TMP_DIR, "cn_region.csv"))
```

```r
cn_ready <- 
cn |>
  mutate(cn_region = paste0("cn_chr", chromosome, "_", chromosomeBand)) |>
  select(-chromosome, -chromosomeBand) |>
  spread( cn_region, cn ) 
```

```r
cn_ready_simple <- 
  cn |>
  mutate(pq = ifelse(grepl("p", chromosomeBand), "p", "q"),
         cn_region_pq = paste0("cn_simple_chr_", chromosome, "_", pq)) |>
  select(-chromosome, -chromosomeBand) |>
  group_by( sampleId, cn_region_pq ) |> 
  summarise( cn = mean(cn)) |>
  spread( cn_region_pq, cn ) 
```

#### 1 - Send processed data

```r
fwrite(cn_ready, paste0(READY_DIR, "cn_ready.csv"))
fwrite(cn_ready_simple, paste0(READY_DIR, "cn_simple_ready.csv"))
```
