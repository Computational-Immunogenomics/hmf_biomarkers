#### 0 - Helpers

```r
SOM_DIR <- paste0(I_DIR, "somatic/")
```

```r
patients <- list.files(SOM_DIR)
```

#### 1 - Collect Purities

```r
purities <- list()
system.time(
for( i in patients){
    i_file <- get_fp(i, type = "purity")
    if(file.exists(i_file)){
        purities[[i]] <- reader(i_file, i)
    }
})
```

```r
fwrite( do.call("rbind", purities), paste0(TMP_DIR, "purities.csv"))
```

#### 2 - Collect Drivers

```r
drivers <- list()
system.time(
for( i in patients){
    i_file <- get_fp(i, type = "drivers")
    if(file.exists(i_file)){
        drivers[[i]] <- reader(i_file, i)
    }
})
```

```r
fwrite( do.call("rbind", drivers), paste0(TMP_DIR, "drivers.csv"))
```

#### 3 - Collect Copy Number

```r
cns <- list()
system.time(
for( i in patients){
    i_file <- get_fp(i, type = "cnv_gene")
    if(file.exists(i_file)){
        cns[[i]] <- 
          reader(i_file, i) |> 
            group_by(sampleId, chromosome, chromosomeBand) |> 
            summarise(cn = mean(maxCopyNumber), cn_minor_allele = mean(minMinorAlleleCopyNumber)) |> 
            ungroup()
    }
})
```

```r
fwrite( do.call("rbind", cns), paste0(TMP_DIR, "cn_region.csv"))
```
